```python
def calculate_AQI(row):
    pm25 = (row['PM2.5'] / 35) * 100
    pm10 = (row['PM10'] / 50) * 100
    so2 = (row['SO2'] / 150) * 100
    no2 = (row['NO2'] / 100) * 100
    co = (row['CO'] / 200) * 100
    o3 = (row['O3'] / 160) * 100
    return max(pm25, pm10, so2, no2, co, o3)

df1['AQI'] = df1.apply(calculate_AQI, axis=1)
```

```python
def label_AQI(aqi):
    if aqi <= 50:
        return 'Excellent'
    elif aqi <= 100:
        return 'Good'
    elif aqi <= 150:
        return 'Slightly Polluted'
    elif aqi <= 200:
        return 'Lightly Polluted'
    elif aqi <= 250:
        return 'Moderately Polluted'
    elif aqi <= 300:
        return 'Heavily Polluted'
    else:
        return 'Severely Polluted'

df1['AQI2'] = df1['AQI'].apply(label_AQI)
```

```python
df1.dtypes
```

```python
from sklearn.model_selection import train_test_split
X = df1[['year', 'month', 'day', 'hour', "PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP", "RAIN", "wd2", "WSPM"]]
y = df1['AQI2']
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC

# Example for training and testing a Support Vector Classifier
model = SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='micro')
recall = recall_score(y_test, y_pred, average='micro')
f1 = f1_score(y_test, y_pred, average='micro')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.model_selection import cross_val_score

# Example of cross-validation using 5 folds
scores = cross_val_score(model, X, y, cv=5)
print("Cross-Validation Scores:", scores)
print("Average Accuracy:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

# Example of hyperparameter tuning using grid search
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
print('Best parameters', best_params)
```

```python
print('Best model', best_model)
```

```python
y_pred = best_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
